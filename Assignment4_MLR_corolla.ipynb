model fitting

```python
import statsmodels.formula.api as smf
model = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data = data).fit()
print('Value of  R Square is' ,model.rsquared)
```

```python
model.params
```

```python
#p vlaues and t values
print(model.pvalues, '\n\n' , model.tvalues)
```

```python
model.summary()
```

from above summary table we can see that the doors is having high correlation. so can try by eliminating it

as there are two variables multi corelated lets see by building a simple linear regresion models for those

```python
sl_cc=smf.ols('Price~cc',data = data).fit()
sl_cc.summary()
```

```python
sl_doors=smf.ols('Price~Doors',data = data).fit()
sl_doors.summary()
```

```python
ml_cc_door =smf.ols('Price~cc+Doors',data = data).fit()
ml_cc_door.summary()
```

variation infliaction factor

```python
vi_cc = smf.ols('cc~Age_08_04+KM+HP+Doors+Gears+Quarterly_Tax+Weight',data = data).fit().rsquared
vif_cc = 1/(1-vi_cc)
vi_door = smf.ols('Doors~cc+Age_08_04+KM+HP+Gears+Quarterly_Tax+Weight',data = data).fit().rsquared
vif_door = 1/(1-vi_door)
d1 = {'variable':['cc','Doors'],'vif value':[vif_cc,vif_door]}
data_vif = pd.DataFrame(d1)
data_vif
```

checking for normality

QQ plot

```python
import statsmodels.api as sm
qq_plot = sm.qqplot(model.resid,line="q")
plt.title('normal qq plot of residuals')
plt.show
```

residual plot for homoscadasity

```python
def get_standardized_values(vals):
  return (vals-vals.mean()/vals.std())
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),get_standardized_values(model.resid))
plt.show()
```
